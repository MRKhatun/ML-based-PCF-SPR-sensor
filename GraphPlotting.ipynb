```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline

df = df = pd.read_excel("phase_matching[tg=50nm.r=0.5um,p=2um and na=1.36].xlsx") 

# Drop rows with NaN or infinite values in any column
df = df.replace([np.inf, -np.inf], np.nan).dropna()

if df.empty:
    print("Error: DataFrame is empty after loading and cleaning. Please check your Excel file.")
else:
    # Smoothing function
    def smooth(x, y, num_points=300):
        x_new = np.linspace(x.min(), x.max(), num_points)
        spl = make_interp_spline(x, y, k=3)
        y_new = spl(x_new)
        return x_new, y_new

    # Figure setup
    fig, ax1 = plt.subplots(figsize=(6,4))

    # Primary y-axis: Confinement loss
    ax1.set_xlabel('Wavelength (µm)', fontsize=14)
    ax1.set_ylabel('Confinement loss (dB/cm)', fontsize=14)

    x_smooth, y_smooth = smooth(df['wl'], df['Confinement loss'])
    line1, = ax1.plot(x_smooth, y_smooth, color='#1f77b4', linewidth=2.2, linestyle='-', label='Confinement loss')

    ax1.tick_params(axis='both', labelsize=14)

    # Secondary y-axis: Effective refractive index
    ax2 = ax1.twinx()
    ax2.set_ylabel('Effective Refractive Index', fontsize=14)

    x_smooth_core, y_smooth_core = smooth(df['wl'], df['neff y core (real)'])
    x_smooth_spp, y_smooth_spp = smooth(df['wl'], df['neff y spp (real)'])

    line2, = ax2.plot(x_smooth_core, y_smooth_core, color='#ff7f0e', linewidth=2.2, linestyle='--', label='Core mode')
    line3, = ax2.plot(x_smooth_spp, y_smooth_spp, color='#2ca02c', linewidth=2.2, linestyle=':', label='SPP mode')

    ax2.tick_params(axis='both', labelsize=12)

    # Combine legends
    ax1.legend(handles=[line1, line2, line3], loc='upper right', fontsize=12, frameon=False)

    plt.tight_layout()
    plt.show()
```
